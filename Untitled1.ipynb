## User based Collaborative recommender system
**This system suggests 5 movies most similar to a given movie based on user ratings across the database**

```python
movies_df.head()
```

```python
rating_df.head()
```

```python
df = pd.merge(rating_df,movies_df,on='movieId')
df.head()
```

```python
combine_movie_rating = df.dropna(axis = 0, subset = ['title'])
movie_ratingCount = (combine_movie_rating.groupby(by = ['title'])['rating'].count().reset_index().rename(columns = {'rating': 'totalRatingCount'})
[['title', 'totalRatingCount']] )
movie_ratingCount.head()
```

```python
rating_with_totalRatingCount = combine_movie_rating.merge(movie_ratingCount, left_on = 'title', right_on = 'title', how = 'left')
rating_with_totalRatingCount.head()
```

```python
popularity_threshold = 10
rating_popular_movie= rating_with_totalRatingCount.query('totalRatingCount >= @popularity_threshold')
rating_popular_movie.head()
```

```python
rating_popular_movie.shape
```

```python
##Pivot matrix

movie_features_df=rating_popular_movie.pivot_table(index='title',columns='userId',values='rating').fillna(0)
movie_features_df.head()
```

```python
from scipy.sparse import csr_matrix

movie_features_df_matrix = csr_matrix(movie_features_df.values)

from sklearn.neighbors import NearestNeighbors


model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(movie_features_df_matrix)
```

```python
movie_features_df.head()
```

```python
def collab_rec(title):
    d=dict()
    i=0
    for index,row in movie_features_df.iterrows():
        d[index]=i
        i+=1
    query_index = d[title]
    distances, indices = model_knn.kneighbors(movie_features_df.iloc[query_index,:].values.reshape(1, -1), n_neighbors = 6)
    movie_features_df.iloc[query_index,:]
    for i in range(0, len(distances.flatten())):
        if i == 0:
            print('Collaborative user based Recommendations for {0}:\n'.format(movie_features_df.index[query_index]))
        else:
            print('{0}: {1}, with distance of {2}:'.format(i, movie_features_df.index[indices.flatten()[i]], distances.flatten()[i]))

   
```

# Content Based Recommender system#

```python
#https://www.kaggle.com/tmdb/tmdb-movie-metadata
credits = pd.read_csv("tmdb_5000_credits.csv")
movies = pd.read_csv("tmdb_5000_movies.csv")
```

```python
credits.rename(index=str, columns={"movie_id": "id"}, inplace=True)
movies_df_merge = movies.merge(credits, on='id')
movies_df_merge.head()
```

```python
movies_df_merge.drop(columns=['homepage', 'title_x', 'title_y', 'status','production_countries'],inplace=True)
movies_df_merge.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer


tfv = TfidfVectorizer(min_df=3,  max_features=None, 
            strip_accents='unicode', analyzer='word',token_pattern=r'\w{1,}',
            ngram_range=(1, 3),
            stop_words = 'english')

# Filling NaNs with empty string
movies_df_merge['overview'] = movies_df_merge['overview'].fillna('')
```

```python
# Fitting the TF-IDF on the 'overview' text
tfv_matrix = tfv.fit_transform(movies_df_merge['overview'])
tfv_matrix
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel

# Compute the sigmoid kernel
sig = sigmoid_kernel(tfv_matrix, tfv_matrix)
sig[0]
```

```python
# Reverse mapping of indices and movie titles
indices = pd.Series(movies_df_merge.index, index=movies_df_merge['original_title']).drop_duplicates()
```

```python
indices
```

```python
def give_rec(title, sig=sig):
    # Get the index corresponding to original_title
    idx = indices[title]

    # Get the pairwsie similarity scores 
    sig_scores = list(enumerate(sig[idx]))

    # Sort the movies 
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)

    # Scores of the 10 most similar movies
    sig_scores = sig_scores[1:11]

    # Movie indices
    movie_indices = [i[0] for i in sig_scores]

    # Top 10 most similar movies
    return movies_df_merge['original_title'].iloc[movie_indices]
```
